```python
import torch
import librosa
import time
import soundfile as sf
import IPython.display as ipd
from model import DCCRN
```

```python
model = torch.load('dccrn.pt').eval()
```

```python
def inference(wav, model):
    device = torch.device('cpu')
    noisy, sr = librosa.load(wav, sr=16000)
    noisy_tensor = torch.tensor([noisy]).to(device)
    clean_array = model(noisy_tensor).squeeze().detach().numpy()
    sf.write(wav[:wav.rfind('.')] + '_processed.wav', clean_array, 16000, format='wav')
```

```python
start = time.time()
inference('./test/keyboard.wav', model)
print(time.time() - start)
```
